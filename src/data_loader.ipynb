```python
%%writefile data_loader.py
import yfinance as yf
import pandas as pd

def get_market_data():
    # BZ=F is Brent Crude, CL=F is WTI, RB=F is Gasoline
    tickers = {
        "Brent_Crude": "BZ=F",
        "WTI_Crude": "CL=F",
        "Gasoline_Futures": "RB=F"
    }
    
    data = yf.download(list(tickers.values()), period="2y", interval="1d")
    
    # Cleaning the multi-index columns
    df = data['Close'].rename(columns={v: k for k, v in tickers.items()})
    return df

if __name__ == "__main__":
    df = get_market_data()
    print("Successfully fetched market data:")
    print(df.tail())
```
